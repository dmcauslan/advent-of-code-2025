```python
import utils
import itertools
```

```python
input1 = utils.read_input_file("4", "1")
```

```python
def find_paper1(inputs):

    def count_if_roll(row_index, col_index):
        if row_index < 0 or row_index > len(inputs) - 1:
            return 0
        if col_index < 0 or col_index > len(inputs[0]) - 1:
            return 0
        if inputs[row_index][col_index] == '@':
            return 1
        return 0
    
    def count_adjacent_rolls(row_index, col_index):
        result = 0
        for (row_offset, col_offset) in itertools.product([-1, 0, 1], [-1, 0, 1]):
            if row_offset == 0 and col_offset == 0:
                continue
            result += count_if_roll(row_index + row_offset, col_index + col_offset)
        return result

    result = 0
    for row_index, row in enumerate(inputs):
        for col_index, value in enumerate(row):
            if value == '@':
                if count_adjacent_rolls(row_index, col_index) < 4:
                    result += 1    
    return result


def find_paper2(inputs):
    def count_if_roll(row_index, col_index):
        if row_index < 0 or row_index > len(inputs) - 1:
            return 0
        if col_index < 0 or col_index > len(inputs[0]) - 1:
            return 0
        if inputs[row_index][col_index] == '@':
            return 1
        return 0
    
    def count_adjacent_rolls(row_index, col_index):
        result = 0
        for (row_offset, col_offset) in itertools.product([-1, 0, 1], [-1, 0, 1]):
            if row_offset == 0 and col_offset == 0:
                continue
            result += count_if_roll(row_index + row_offset, col_index + col_offset)
        return result

    result = 0
    next_board = ""
    for row_index, row in enumerate(inputs):
        for col_index, value in enumerate(row):
            if value == '@':
                if count_adjacent_rolls(row_index, col_index) < 4:
                    result += 1    
                    next_board += "x"
                else:
                    next_board += "@"
            else:
                next_board += "."
        next_board += "\n"
    return result, next_board.strip().split("\n")

def count_all_moveable_rolls(inputs):
    board = inputs
    count = 0
    while True:
        # print("\n".join(board))
        rolls_moved, next_board = find_paper2(board)
        count += rolls_moved
        board = next_board
        if rolls_moved == 0:
            break
    return count
```

```python
sample_text = """
..@@.@@@@.
@@@.@.@.@@
@@@@@.@.@@
@.@@@@..@.
@@.@@@@.@@
.@@@@@@@.@
.@.@.@.@@@
@.@@@.@@@@
.@@@@@@@@.
@.@.@@@.@.
"""
sample_input = [r for r in sample_text.split("\n") if r]
assert find_paper1(sample_input) == 13
assert count_all_moveable_rolls(sample_input) == 43
```
