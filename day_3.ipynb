```python
def find_max_part_1(input_str):
    values = [int(x) for x in input_str]
    first_max_value = max(values[:-1]) # Ignore last value since we need two digits
    first_max_index = input_str.index(str(first_max_value))
    second_max_value = max(values[first_max_index+1:])
    return int(f"{first_max_value}{second_max_value}")

def find_max_part_2_recursively(input_str):
    # Waaaay too slow
    print(input_str)
    def find_max(digits):
        print(digits)
        if len(digits) == 11:
            return int("".join(digits))
        max_value = 0
        for i in range(len(digits)):
            digits_copy = digits[:]
            digits_copy.pop(i)
            max_value = max(max_value, find_max(digits_copy))
        return max_value
    
    values = [int(x) for x in input_str]
    first_max_value = max(values[:-11]) # Ignore last value since we need 12 digits
    first_max_index = input_str.index(str(first_max_value))
    remaining_numbers = list(input_str[first_max_index+1:]) # Ignore the max value - we'll add it back later.
    largest_remainder = find_max(remaining_numbers)
    return int(f"{first_max_value}{largest_remainder}")
    
def find_max_part_2_rising_tide(input_str):
    # Eliminates the lowest value 1, 2, 3 until we have fewer than 11 characters remaining
    # But it doesn't work - I think it's possible it prefers a number like 945 over 961.
    def drop_digits(digits, i):
        if len(digits) == 11:
            return "".join(digits)
        remainder = [d for d in digits if d != str(i)]
        if len(remainder) < 11:
            raise ValueError(f"Too short {digits}")
        try:
            return drop_digits(remainder, i+1)
        except:
            # The next stage was too far, so remove the N most important values with that number until we get to 11
            needed_to_remove = len(remainder) - 11
            while needed_to_remove:
                # print(remainder, i+1, needed_to_remove)
                remainder.remove(str(i+1))
                needed_to_remove -= 1
            return "".join(remainder)
    
    values = [int(x) for x in input_str]
    first_max_value = max(values[:-11]) # Ignore last value since we need 12 digits
    first_max_index = input_str.index(str(first_max_value))
    remaining_numbers = list(input_str[first_max_index+1:]) # Ignore the max value - we'll add it back later.
    largest_remainder = drop_digits(remaining_numbers, 0)
    result = int(f"{first_max_value}{largest_remainder}")
    return result

def find_max_part_2_successive_maxes(input_str):
    # Find the next max number in the remaining digits
    needed_digits = 12
    digits = []
    last_index = 0
    values = [int(x) for x in input_str]
    for i in range(needed_digits):
        end = -needed_digits + i + 1
        if end < 0:
            chunk = values[last_index:end]
        else:
            chunk = values[last_index:]
        # print(digits, last_index, i, chunk)
        next_highest = max(chunk)
        digits.append(str(next_highest))
        last_index = input_str.index(str(next_highest),last_index) + 1
    return int("".join(digits))
```

```python
assert find_max_part_1("1234") == 34
assert find_max_part_2_successive_maxes("234234234234278") == 434234234278
assert find_max_part_2_successive_maxes("987654321111111") == 987654321111
assert find_max_part_2_successive_maxes("811111111111119") == 811111111119
assert find_max_part_2_successive_maxes("818181911112111") == 888911112111
```
