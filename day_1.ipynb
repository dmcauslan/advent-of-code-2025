```python
import utils
```

```python
inputs = [(i[0], int(i[1:])) for i in utils.read_input_file("1", "1")]
```

```python
def process_input_1(current_position, direction, count) -> int:
    if direction == 'R':
        current_position = (current_position + count) % 100
    else:
        current_position = (current_position - count) % 100
    if current_position == 0:
        return 1
    else:
        return 0

def process_input_2(current_position, direction, count) -> tuple[int, int]:
    zeroes = 0

    if direction == 'R':
        if current_position != 0:
            clicks_to_100 = 100 - current_position
            if clicks_to_100 > count:
                return current_position + count, 0
            zeroes += 1
            count -= clicks_to_100
        zeroes += count // 100
        return count % 100, zeroes
    else:
        if current_position != 0:
            clicks_to_zero = current_position
            if clicks_to_zero > count:
                return current_position - count, 0
            zeroes += 1
            count -= current_position
        zeroes += count // 100
        return (0 - count) % 100, zeroes

def process_inputs(inputs) -> int:
    zero_count = 0
    position = 50
    for (direction, count) in inputs:
        new_position, zeroes = process_input_2(position, direction, count)
        position = new_position
        zero_count += zeroes
    return zero_count
```

```python
process_inputs(inputs)
```

```python
# tests
assert process_input_2(0, "R", 100) == (0, 1)
assert process_input_2(0, "R", 1000) == (0, 10)
assert process_input_2(1, "R", 999) == (0, 10)
assert process_input_2(1, "R", 101) == (2, 1)
assert process_input_2(1, "R", 99) == (0, 1)
assert process_input_2(1, "R", 2) == (3, 0)

assert process_input_2(0, "L", 100) == (0, 1)
```
