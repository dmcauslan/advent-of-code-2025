```python
def find_invalid_ids_part_1(range_str):
    start_str, end_str = range_str.split("-")
    start_int = int(start_str)
    end_int = int(end_str)

    # Strings must be even in order to have repetitions
    # Eg if start is 990, start looking for repetitions at 1000
    if len(start_str) % 2 != 0:
        start_str = "1" + "0" * len(start_str)

    first_half = int(start_str[:len(start_str) // 2])

    while True:
        number = int(f"{first_half}{first_half}")
        if number >= start_int and number <= end_int:
            yield number
        if number > end_int:
            break
        first_half += 1
```

```python
def get_max_input_value():
    max_value = 0
    for r in input_1.split(","):
        start_str, end_str = r.split("-")
        max_value = max(max_value, int(end_str))
    return max_value

MAX_INPUT_VALUE = get_max_input_value()
print(f"{MAX_INPUT_VALUE=}")

def generate_repeated_numbers(block_length):
    start = int("1" + "0" * (block_length - 1))
    max_number_length = len(str(MAX_INPUT_VALUE))
    max_repetitions = max_number_length // block_length
    result = list()
    for r in range(2, max_repetitions + 1):
        counter = start
        while True:
            number = int(str(counter)*r)
            if number > MAX_INPUT_VALUE:
                break
            result.append(number)
            counter += 1
            if len(str(counter)) > block_length:
                break
    return result

def generate_all_repeated_numbers():
    result = list()
    max_block_length = len(str(MAX_INPUT_VALUE)) // 2
    for i in range(1, max_block_length + 1):
        result.extend(generate_repeated_numbers(i))
    return sorted(set(result))

ALL_NUMBERS_WITH_REPETITIONS = generate_all_repeated_numbers()

def find_invalid_ids_part_2(range_str):
    start_str, end_str = range_str.split("-")
    start_int = int(start_str)
    end_int = int(end_str)

    # Dumb filter with iteration. Maybe fast enough with only ~10k elements.
    return [x for x in ALL_NUMBERS_WITH_REPETITIONS if x >= start_int and x <= end_int]
```

```python
assert list(find_invalid_ids_part_1("11-22")) == [11, 22]
assert list(find_invalid_ids_part_1("1188511880-1188511890")) == [1188511885]

assert list(find_invalid_ids_part_2("11-22")) == [11, 22]
assert list(find_invalid_ids_part_2("1188511880-1188511890")) == [1188511885]
```

```python
input_1 = utils.read_input_file("2", "1")[0]
id_ranges = [s for s in input_1.split(",")]
invalid_id_lists = list(list(find_invalid_ids_part_1(i)) for i in id_ranges)
result = 0
for invalid_ids in invalid_id_lists:
    result += sum(invalid_ids)
result
```
